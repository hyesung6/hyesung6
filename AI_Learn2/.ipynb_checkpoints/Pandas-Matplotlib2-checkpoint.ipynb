# 시각화

### Matplotlib - 기본 그래프 도구

```python
'''
데이터를 차트나 플롯으로 시각화 하는 패키지
line plot, scatter plot, bar plot, histogram 등
다양한 비쥬얼라이제이션(시각화) 기능을 제공

Matplotlib의 그림은 Figure, Axes, Axis 객체가 있다.
Figure : 플롯이 그려지는 캔버스
Axes : 하나의 그림(플롯)
Axis : x, y... 축 ( Axes객체는 2개 이상의 Axis 객체를 포함한다. )

'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib는 인코딩 문제로 한글이 인식 안됨
from matplotlib import font_manager, rc
font_path = 'data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

# 선그래프

```python
df = pd.read_excel('data/시도별 전출입 인구수.xlsx')
df.head()
```

```python
df.info()
```

```python
# 전출지별 NaN값을 앞 데이터 채움

df = df.fillna(method='ffill')       # ffill : front fill 앞에 오는 항목으로 채우기
df.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul.head()
```

```python
# 전출지 삭제

df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.head()
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
df_seoul.info


# 데이터 분석 쪽으로 공부를 계속 하려면
# 이정도 가공은 할 수 있어야 한다...
```

```python
# 서울에서 경기도로 이동한 인구 데이터 변화
ggd = df_seoul.loc['경기도']

# 그림 사이즈 지정 (x, y)
plt.figure(figsize=(30, 5))

# plot 함수에 입력
plt.plot(ggd.index, ggd.values)

# 1995년에 서울의 전출인구가 가장 높은데, 해당년도에 무슨 일이 있었는지 찾아보는 것이 그래프를 의미있게 만들 수 있다.

# 차트 제목 추가
plt.title('seoul --> GYEONGGI pop')

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show()
```

# 면적 그래프

```python
# 서울에서 충남, 경북, 강원, 전남으로 이동한 인구데이터만 선택하기

col_years = list(map(str, range(1970, 2018)))
col_years
```

```python
df_2 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_2
```

```python
df_2 = df_2.transpose()     # transpose() : T와 같음. 열과 행 바꿈
df_2
```

```python
# 스타일 목록
# plt.style.available

# 스타일 서식 지정
# plt.style.use('ggplot')

# 그림 사이즈 늘리기
# plt.figure(firgsize=(x, y))

# x, y축 데이터를 plot 함수에 입력
df_2.plot(kind='area', stacked=False, alpha=0.4, figsize=(20,10))          
# scatter : 선그래프    area : 면적그래프      bar : 막대그래프
# barh : 가로 막대그래프    hist : 히스토그램   
# stacked : 데이터의 누적 유무(T/F)
# alpha : 0일수록 밝음, 1일수록 어두움.

plt.title('서울 ---> 타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

# plt.legend() 범례 지정하기

plt.show()
# *** 이 그래프는 스택 False ***
```

```python
df_2.plot(kind='area', stacked=True, alpha=0.4, figsize=(20,10)) 

plt.title('서울 ---> 타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
# *** 이 그래프는 스택 True ***  
```

#  막대 그래프

```python
'''
막대 높이의 상대적 길이 차이를 통해 값의 크고 작음을 설명하고자 할 때
세로형 막대그래프 - 시간적으로 차이가 나는 두점에서 데이터 값의 차이를 잘 설명해주는 그래프
                 - 시계열 데이터에 적합              
'''
```

```python
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_3 = df_3.transpose()

df_3.plot(kind='bar', figsize=(20,10), width=0.5, color=['blue', 'red', 'purple', 'green']) 

plt.title('서울 ---> 타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
'''
가로형 막대 그래프 : 각 변수 사이의 값의 크기 차이를 설명하는 데 적합
'''

df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_4['합계'] = df_4.sum(axis=1)

# 가장 큰 값부터 정렬
df_total =df_4[['합계']].sort_values(by='합계', ascending=True)

df_4.plot(kind='barh', figsize=(20,10), width=0.5, color='skyblue') 

plt.title('서울 ---> 타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

# 히스토그램

```python
'''
변수가 하나인 데이터의 빈도수를 막대형으로 표현함
x축을 같은 크기의 여러 구간으로 나누고 각 구간에 속하는 데이터 값의 개수(빈도)를 y축에 표시
'python' 과목 ~59점이 30퍼, 60~70~ 20퍼, 71~80이 20퍼, 81~90이 20퍼, 91~100이 10퍼라는 식으로..
'''
```

#  산점도

```python
'''
서로 다른 두 변수 사이의 관계를 나타낼 때 사용
(일단 그래프로 두 변수를 그리고 그 안에서 의미를 찾아내야할 때)
이 때 각 변수는 연속되는 값을 갖는다. 일반적으로 정수, 실수형의 값
2개의 연속 변수를 각 x,y축에 놓고 데이터값이 위치하는 (x,y) 좌표를 찾아 점으로 표기
'''

plt.style.use('default')

df = pd.read_csv('data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

df.plot(kind='scatter', x='weight', y='mpg', c='coral', s=10, figsize=(14, 5))

plt.title('Scatter plot - mpg || weight')

plt.show()


# 해당 산점도를 통해 차중(x)과 연비(y) 간의 상관관계에 대해 추측해볼 수 있다.
```

```python
plt.style.use('default')

df = pd.read_csv('data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

# 실린더 개수의 상대적 비율
cylinder_ratio = df.cylinders / df.cylinders.max() * 300

df.plot(kind='scatter', x='weight', y='mpg', c='coral', s=cylinder_ratio, figsize=(14, 5), alpha=0.4)

plt.title('Scatter plot - mpg || weight')

plt.savefig('image/scatter-bubble.png')
plt.savefig('image/scatter-bubble2.png', transparent=True)

plt.show()


# 점의 크기에 변화를 주면 비눗방울처럼 보인다고 해서 버블차트라고 함

# 결과 그래프를 저장하려면 plt.show() 전에
# plt.savefig('저장할파일')
```

#  파이 차트

```python
plt.style.use('default')

df = pd.read_csv('data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

# 제조국가(origin) 열에 대한 파이 차트 그리기
# 파이 차트는 백분율을 나타내야 하므로, 전체 중 몇 개에 해당하는지 연산을 미리 해야함

df['count'] = 1
df_origin = df.groupby('origin').sum()
df_origin

```

```python
# 제조국을 실제 지역명으로 변경
df_origin.index = ['USA', 'EU', 'JAPAN']

# 제조국가 열에 대한 파이차트 그리기
df_origin['count'].plot(kind='pie', figsize=(14, 5), autopct='%1.1f%%', startangle=90, colors=['bisque', 'pink', 'chocolate'])

plt.title('origin', size=20)
plt.axis('equal')      # 비율이 큰 부분이 찌그러지게 나오게 되어있으므로, 원에 가깝게 보정을 해줌
plt.legend(labels=df_origin.index, loc='upper right')

plt.savefig('image/oring-pie.png')
plt.show
```
