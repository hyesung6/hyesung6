```python
'''
데이터를 차트나 플롯으로 시각화 하는 패키지
line plot, scatter plot, bar plot, histogram 등
다양한 비쥬얼라이제이션(시각화) 기능을 제공

Matplotlib의 그림은 Figure, Axes, Axis 객체가 있다.
Figure : 플롯이 그려지는 캔버스
Axes : 하나의 그림(플롯)
Axis : x, y... 축 ( Axes객체는 2개 이상의 Axis 객체를 포함한다. )

'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib는 인코딩 문제로 한글이 인식 안됨
from matplotlib import font_manager, rc
font_path = 'data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

# 선그래프

```python
df = pd.read_excel('data/시도별 전출입 인구수.xlsx')
df.head()
```

```python
df.info()
```

```python
# 전출지별 NaN값을 앞 데이터 채움

df = df.fillna(method='ffill')       # ffill : front fill 앞에 오는 항목으로 채우기
df.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul.head()
```

```python
# 전출지 삭제

df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.head()
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
df_seoul.info


# 데이터 분석 쪽으로 공부를 계속 하려면
# 이정도 가공은 할 수 있어야 한다...
```

```python
# 서울에서 경기도로 이동한 인구 데이터 변화
ggd = df_seoul.loc['경기도']

# 그림 사이즈 지정 (x, y)
plt.figure(figsize=(30, 5))

# plot 함수에 입력
plt.plot(ggd.index, ggd.values)

# 1995년에 서울의 전출인구가 가장 높은데, 해당년도에 무슨 일이 있었는지 찾아보는 것이 그래프를 의미있게 만들 수 있다.

# 차트 제목 추가
plt.title('seoul --> GYEONGGI pop')

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show()
```
