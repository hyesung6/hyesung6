# 데이터 살펴보기 

```python
# 데이터 살펴보기
# auto-mpg.csv
# 데이터의 출처와 데이터 항목의 의미를 알아야 한다.
# 유의미한 데이터가 중요
```

```python
'''
# UCI 머신러닝 저장소에서 제공하는 자동차 연비(auto-mpg) 데이터 세트

   속성         :   데이터 상세(범위)
1. mpg          :   continuous(연속값)                  연비
2. cylinders    :   multi-valued discrete(이산값)       실린더수
3. displacement :   continuous                          배기량
4. horsepower   :   continuous                          출력
5. weight       :   continuous                          차중
6. acceleration :   continuous                          가속능력
7. model year   :   multi-valued discrete               출시년도
8. origin       :   multi-valued discrete               제조국
9. car name     :   string (unique for each instance)   모델명

'''
```

```python
# csv 읽어오기

df = pd.read_csv('data/auto-mpg.csv', header=None)
df
```

```python
# 일부 데이터를 확인 : head(), tail()      # default : 5개 행
# 데이터 간략하게 확인하고 싶을 때
```

```python
print(df.head())
print('-'*100)
print(df.tail())
```

```python
# 컬럼이름 지정해서 데이터 프레임 채우기

# 1. mpg: continuous
# 2. cylinders: multi-valued discrete
# 3. displacement: continuous
# 4. horsepower: continuous
# 5. weight: continuous
# 6. acceleration: continuous
# 7. model year: multi-valued discrete
# 8. origin: multi-valued discrete
# 9. car name: string (unique for each instance)
    
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
df.head()
```
