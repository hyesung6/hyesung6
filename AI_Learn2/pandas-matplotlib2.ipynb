# 시각화
## Matplotlib - 기본 그래프 도구

```python
'''
# 데이터를 차트(chart)나 플롯(plot)으로 시각화 하는 패키지
# line plot, scatter plot, bar plot, histogram 등등 여러 다양한 시각화 기능을 제공

# Matplotlib를 통해서 그려내는 그림은 Figure 객체, Axes 객체, Axis 객체
# Figure : 플롯이 그려지는 캔버스 (한 개 이상의 Axes 객체를 포함한다)
# Axes : 하나의 그림(플롯)
# Axis : Axes 객체가 2개 이상의 Axis 객체를 포함한다.

'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib 한글 폰트 인식 안됨
from matplotlib import font_manager, rc
font_path = 'data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

## 선그래프

```python
df = pd.read_excel('data/시도별 전출입 인구수.xlsx')
df.head()
```

```python
df.info()
```

```python
# 전출지별 NaN값을 앞 데이터로 채움
df = df.fillna(method='ffill')
df.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출 
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul.head()
```

```python
# 전출지 삭제
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.head()
```

```python
df_seoul.rename({'전입지별':'전입지'},  axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.info
```

```python
# 서울에서 경기도로 이동한 인구 데이터 변화
ggd = df_seoul.loc['경기도']

# 그림 사이즈 지정 (가로 x 인치, 세로 y 인치)
plt.figure(figsize=(30, 5))

# plot 함수에 입력
plt.plot(ggd.index, ggd.values)
```

```python
# 그림 사이즈 지정 (가로 x 인치, 세로 y 인치)
plt.figure(figsize=(30, 5))

# plot 함수에 입력
plt.plot(ggd)
```

```python
# 그림 사이즈 지정 (가로 x 인치, 세로 y 인치)
plt.figure(figsize=(30, 5))

# plot 함수에 입력
plt.plot(ggd)

# 차트 제목 추가
plt.title('seoul ==> GYEONGGI pop')

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show()
```

## 면적 그래프

```python
# 서울에서 충남, 경북, 강원, 전남으로 이동한 인구데이터만 선택
col_years = list(map(str, range(1970, 2018)))
col_years
```

```python
df_2 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_2
```

```python
df_2 = df_2.transpose()  # transpose() : T
df_2
```

```python
# 스타일 리스트 
#plt.style.available

# 스타일 서식 지정
plt.style.use('ggplot')

# x, y 축 데이터를 plot 함수에 입력
df_2.plot(kind='area', stacked=False, alpha=0.5, figsize=(20, 10))

plt.title('서울 --->  타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
# 스타일 서식 지정
plt.style.use('ggplot')

# x, y 축 데이터를 plot 함수에 입력
df_2.plot(kind='area', stacked=True, alpha=0.5, figsize=(20, 10))

plt.title('서울 --->  타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

## 막대 그래프

```python
'''
# 막대 높이의 상대적 길이 차이를 통해 값의 크고 작음을 설명하고자 할 때
# 세로형 막대그래프 - 시간적으로 차이가 나는 두 점에서 데이터 값의 차이를 잘 설명해주는 그래프
                   - 시계열 데이터에 적합
'''
```

```python
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_3 = df_3.transpose()

# 스타일 서식 지정
plt.style.use('ggplot')

# x, y 축 데이터를 plot 함수에 입력
df_3.plot(kind='bar', figsize=(40, 10), width=0.7, color=['orange', 'green', 'skyblue', 'red'])

plt.title('서울 --->  타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
'''
# 가로형 막대 그래프 : 각 변수 사이의 값의 크기 차이를 설명하는데 적합
'''
```

```python
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]

df_3['합계'] = df_3.sum(axis=1)

# 가장 큰 값부터 정렬
df_total = df_3[['합계']].sort_values(by='합계', ascending=True)


# 스타일 서식 지정
plt.style.use('ggplot')

# x, y 축 데이터를 plot 함수에 입력
df_total.plot(kind='barh', figsize=(20, 10), width=0.7, color='cornflowerblue')

plt.title('서울 --->  타시도 인구 이동', size=30)
plt.ylabel('이동인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

## 산점도 

```python
'''
# 서로 다른 두 변수 사이의 관계를 나타냄
# 이 때 각 변수는 연속되는 값을 갖는다. 일반적으로 정수형 또는 실수형 값
# 2개의 연속 변수를 각각 x축, y축에 놓고 데이터 값이 위치하는 (x,y) 좌표를 찾아서 점으로 표기
'''
```

```python
# 스타일 서식 지정
plt.style.use('default')

# 데이터 불러오기
df = pd.read_csv('data/auto-mpg.csv', header=None)

# 컬럼 이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

# 연비(mpg)와 차중(weight) 열에 대한 산점도
df.plot(kind='scatter', x='weight', y='mpg', c='coral', s=10, figsize=(14, 5))

plt.title('Scatter plot - mpg || weight')
plt.show()
```

```python
# 스타일 서식 지정
plt.style.use('default')

# 데이터 불러오기
df = pd.read_csv('data/auto-mpg.csv', header=None)

# 컬럼 이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

# 실린더 개수의 상대적 비율
cylinder_rate = df.cylinders / df.cylinders.max() * 300

# 연비(mpg)와 차중(weight) 열에 대한 산점도
df.plot(kind='scatter', x='weight', y='mpg', c='coral', s=cylinder_rate, figsize=(14, 5), alpha=0.4)

plt.title('Scatter plot - mpg || weight')


# 점의 크기에 변화를 주면 비눗방울 처럼 보인다고 해서 버블차트라고 함.

# 결과 그래프 저장
plt.savefig('img/scatter-bubble.png')
plt.show()
```

## 파이 차트

```python
# 스타일 서식 지정
plt.style.use('default')

# 데이터 불러오기
df = pd.read_csv('data/auto-mpg.csv', header=None)

# 컬럼 이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

df.head()
```

```python
# 제조국가 열에 대한 파이 차트 그리기

# origin 열을 기준으로 그룹화, 합계 연산 -> origin 별 몇개가 있는지 카운트 할 수 있도록 처리
df['count'] = 1
df_origin = df.groupby('origin').sum()
df_origin
```

```python
# 제조국을 실제 지역명으로 변경
df_origin.index = ['USA', 'EU', 'JAPAN']

# 제조국가 열에 대한 파이차트 그리기
df_origin['count'].plot(kind='pie', figsize=(14, 5), autopct='%1.1f%%', startangle=90, colors=['chocolate', 'bisque', '#aaa'])

plt.title('origin', size=20)
plt.axis('equal')  # 파이 차트 비율을 같게 => 원에 가깝게 조정
plt.legend(labels=df_origin.index, loc='upper right')
plt.show()
```
