## 지도에 마커 표시하기

```python
import pandas as pd
import folium
```

```python
# 대학교 리스트를 읽어오기
df = pd.read_excel('data/서울지역 대학교 위치.xlsx')
df.columns = ['name', '위도', '경도']
df.head()
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(location=[37.56, 126.97], tiles='Stamen Toner', zoom_start=13)

# 대학교 위치 정보를 마커에 표시
for name, lat, lng in zip(df.name, df.위도, df.경도):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)

seoul_map
```

## 지도 영역에 단계 구분도 표시하기

```python
import pandas as pd
import folium
import json
```

```python
# 경기도 인구 변화 데이터 정보 불러오기
file_path = 'data/경기도인구데이터.xlsx'
df = pd.read_excel(file_path, index_col='구분')
df.head()
```

```python
df.columns = df.columns.map(str)
df.columns
```

```python
# 경기도 시군구 경계정보를 가진 geo-json 파일 읽어오기
geo_path = 'data/경기도행정구역경계.json'

geo_data = json.load(open(geo_path, encoding='utf-8'))


# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], tiles='Stamen Terrain', zoom_start=9)
```

```python
# 정보가 필요한 년도 선택해서 그 정보를 지도에 출력
year = '2007'

# Choropleth 클래스로 단계구분도 표시하기
folium.Choropleth(geo_data=geo_data, data=df[year], columns=[df.index, df[year]], 
                 fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.3,
                  threshold_scale=[10000, 10000, 30000, 50000, 700000],
                  key_on='feature.properties.name',
                 ).add_to(g_map)

g_map
```
